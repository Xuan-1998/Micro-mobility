```python
import pandas as pd
```

```python
AustinIntercepts = pd.read_csv('activity-intercepts-Austin.csv')
```

```python
AustinIntercepts['sum'] = AustinIntercepts['Other'] + AustinIntercepts['Shopping'] + AustinIntercepts['Meal'] + AustinIntercepts['SocRec'] + AustinIntercepts['Home'] + AustinIntercepts['Work']
```

```python
AustinIntercepts['Home'].plot(title='Home')
```

```python
AustinIntercepts['Work'].plot(title='Work')
```

```python
AustinIntercepts['Other'].plot(title='Other')
```

```python
AustinIntercepts['Shopping'].plot(title = 'Shopping')
```

```python
AustinIntercepts['Meal'].plot(title = 'Meal')
```

```python
AustinIntercepts['SocRec'].plot(title = 'Social/Recreational')
```

```python
AustinIntercepts['sum'].plot(title='total_including_home and work')
```

```python
AustinIntercepts.loc['sum'] = AustinIntercepts.apply(lambda x : x.sum())
forPie = AustinIntercepts[['Other', 'Shopping', 'Meal', 'SocRec', 'Home', 'Work']]
plot = forPie.T.plot.pie(y='sum', figsize=(15, 25), autopct='%1.1f%%', subplots=True, labels=['Other', 'Shopping', 'Meal', 'SocRec', 'Home', 'Work'])
```
